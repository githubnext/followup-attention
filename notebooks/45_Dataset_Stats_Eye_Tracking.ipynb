# 2. How much time participants approximately spend per task?


```python
# recursively extract the path to all csv ending with "_fixations.csv"
# in the dataset folder

import os
import re

def get_fixation_data(dataset_path: str) -> pd.DataFrame:
    """Extract the fixation data recursively from the dataset folder.

    For each fixation recording we get the:
    - the columsn of the _fixation.csv files plus
    - a column with the name of the folder in which it is contained
    - the full path to the _fixation.csv file
    """
    dfs = []
    for root, dirs, files in os.walk(dataset_path):
        for file in files:
            if file.endswith("_fixations.csv"):
                path_to_store = os.path.join(root, file)
                folder_to_store = os.path.basename(root)
                df = pd.read_csv(path_to_store)
                # keep only if there are data
                if df.shape[0] == 0:
                    continue
                df["user_folder"] = folder_to_store
                df["data_path"] = path_to_store
                # get the name of the column starting with TIME(
                # e.g. TIME(2022/08/10 11:34:07.145)
                # and extract the date and time and store it in a new column
                all_columns = df.columns
                relevant_column = [col for col in all_columns if col.startswith("TIME(")][0]
                date = re.findall(r'\((.*?)\)', relevant_column)[0]
                # convert the date to datetime
                df["date"] = pd.to_datetime(date)
                # for each row add the number of seconds (contained in the relevnat column)
                # to the date column 
                df["date"] = df["date"] + pd.to_timedelta(df[relevant_column], unit="s")
                # rename the relevant column with "seconds since start"
                df = df.rename(columns={relevant_column: "seconds_since_start"})
                # add the dataframe to the list
                dfs.append(df)
    print(f"Found {len(dfs)} fixation files")
    return pd.concat(dfs)

df_fixation_data = get_fixation_data(DATASET_FOLDER)
df_fixation_data.head()
```

```python
# group by data_path and get the max of seconds_since_start
# this is the time spent on the task
df_time_spent = df_fixation_data.groupby(["data_path", "user_folder"])["seconds_since_start"].max().reset_index()
# convert the seconds to minutes
df_time_spent["minutes_spent"] = df_time_spent["seconds_since_start"] / 60
df_time_spent.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_histogram_of(
        column: str, 
        df: pd.DataFrame, 
        latex_variable_name: str = "Empty",
        color: str = "dodgerblue",
        x_label: str = None,
        y_label: str = None):
    """Plot the histogram of the given column."""
    fig, ax = plt.subplots(figsize=(5, 3))
    sns.histplot(
        data=df,
        x=column,
        color=color,
        bins=20,
    )
    mean_val = df[column].mean()
    ax.axvline(mean_val, color="red", linestyle="--", label=f"Mean: {mean_val:.2f}")
    print("\\newcommand{\\mean" + latex_variable_name + "}{" + f"{mean_val:.2f}" + "}")
    median_val = df[column].median()
    ax.axvline(median_val, color="green", linestyle="--", label=f"Median: {median_val:.2f}")
    print("\\newcommand{\\median" + latex_variable_name + "}{" + f"{median_val:.2f}" + "}")
    plt.legend()

    if x_label is not None:
        ax.set_xlabel(x_label)
    if y_label is not None:
        ax.set_ylabel(y_label)
    # show the legend
    ax.set_xlim(0, df[column].max())
    return fig, ax

plot_histogram_of(
    column="minutes_spent",
    df=df_time_spent,
    latex_variable_name="EffectiveTimeSpentPerTaskMinutes",
    x_label="Time spent per question (min)",
    y_label="Number of sessions\n(single question)",
)
```

```python
# group by user_folder and get the sum of all the time spent
# this is the time spent by a user on all the tasks
df_time_spent_user = df_time_spent.groupby(["user_folder"])["minutes_spent"].sum().reset_index()
df_time_spent_user.head()
```

```python
plot_histogram_of(
    column="minutes_spent",
    df=df_time_spent_user,
    latex_variable_name="EffectiveTimeSpentPerUserMinutes",
    color="darkorange",
    x_label="Time spent per participant (min)",
    y_label="Number of participants",
)
```

# 3. how many fixations they had per task?

```python
# group by data_path and get the count of the rows
# this is the number of fixation datapoints
df_fixation_count = df_fixation_data.groupby(["data_path", "user_folder"])["seconds_since_start"].count().reset_index()
# rename the column to fixation_count
df_fixation_count = df_fixation_count.rename(columns={"seconds_since_start": "fixation_count"})
df_fixation_count.head()
```

```python
plot_histogram_of(
    column="fixation_count",
    df=df_fixation_count,
    latex_variable_name="NumberOfFixationsPerTask",
    color="salmon",
    x_label="Number of fixations per question",
    y_label="Number of sessions\n(single question)",
)
```


# 4. how many human participants looked at the same code snippets (regardless of the question)?


```python
HUMAN_ANNOTATIONS = "../data/annotations/2022-09-18/Answer Evaluation - Users - HUMAN_SCORE.csv"
```

```python
df_human_score = pd.read_csv(HUMAN_ANNOTATIONS)

df_human_score = df_human_score[["user", "filename", "score", "unique_data_id"]]
print(len(df_human_score))
replacement_dict = {
    0: "Wrong",
    1: "Partial",
    2: "Correct",
}
df_human_score["score"] = df_human_score["score"].replace(replacement_dict)
df_human_score["task"] = df_human_score["filename"].apply(
    lambda x: x.split("_")[0])
df_human_score["pl"] = df_human_score["filename"].apply(
    lambda x: x.split(".")[-1])
df_human_score["code_snippet"] = df_human_score["task"] + "_" + df_human_score["pl"]
df_human_score["question"] = df_human_score["filename"].apply(
    lambda x: x.split("_")[1].split(".")[0])
df_human_score.head()
```

```python
# count how many participants looked at the same code snippet
df_part_per_code_snippet = df_human_score.groupby(["code_snippet"])["user"].count().reset_index().sort_values(by="user", ascending=False)
# rename to number of participants
df_part_per_code_snippet = df_part_per_code_snippet.rename(columns={"user": "number_of_participants"})
df_part_per_code_snippet.head()
```

```python
fig, ax = plot_histogram_of(
    column="number_of_participants",
    df=df_part_per_code_snippet,
    latex_variable_name="NumberOfParticipantsPerCodeSnippet",
    color="darkgreen",
    x_label="Number of participants",
    y_label="Number of code snippets",
)
import matplotlib.ticker as ticker
# force the x ticks to be integers, every 1
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
# force the x ticks to be integers, every 1
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())

```


# 5. how many human participants looked at the same code snippets with the same question?

```python
# count how many participants looked at the same code snippet (and question)
df_part_per_code_snippet_question = df_human_score.groupby(["code_snippet", "question"])["user"].count().reset_index().sort_values(by="user", ascending=False)
# rename to number of participants
df_part_per_code_snippet_question = df_part_per_code_snippet_question.rename(columns={"user": "number_of_participants"})
df_part_per_code_snippet_question.head()
```
